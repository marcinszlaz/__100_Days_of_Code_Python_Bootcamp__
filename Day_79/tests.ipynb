Figure == canvas, Axes == individual plot, Axis x,y axis
plt.show() dispalys the plot. Plots are displayed automatically, but it's good practise to include it.

```python
# Method first
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.plot(y, x)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plot tests')
plt.show() # Plots are displayed automatically (in jupyter), but it's good practice to include it. (in normal Python interpreter without plt.show() You don't see anything)
```

```python
# Method 2: Object-oriented approach (more control ,recommended for comples plots)
fig, ax = plt.subplots() # Creates a figure and axes obj
# subplots()->Figure, Any
ax.plot(x, y)
ax.plot(y, x)
ax.set_xlabel('X axis') # in ax you add prefix set_(xlabel, ylabel)
ax.set_ylabel('Y axis') # in plt direct xlabel, ylabel...
ax.set_title('Line Plot (Object-Oriented)')
plt.show()
```

```python
# Create some sample data
x = np.linspace(0, 10,100) # generates 100 float digits from 0 to 10
y = np.sin(x) # return sinus for x

# Simple line plot
plt.figure(figsize=(8,4), dpi=100)
plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.yticks(rotation=45)
plt.xticks(rotation=45)
plt.title('Sine Wave')
plt.grid(visible=True, which='major', axis='both')
plt.show()
```

```python
# Multiple lines on the same plot
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

plt.figure(figsize=(10,6)) # Length / height
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='sin(x) * cos(x)')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Multiple Functions')
plt.legend() # Show legend
plt.grid(visible=True, alpha=0.7) # alpha controls transparency
plt.show()
```

```python
x = np.linspace(0, 10, 100)

# Different line styles and markers
plt.figure(figsize=(12,6), dpi=200)
plt.plot(x, x, 'b-', label='Solid line', linewidth=2)
plt.plot(x, x+1, 'r--', label='Dashed line', linewidth=2)
plt.plot(x, x+2, 'g-.', label='Dash-dot line', linewidth=2)
plt.plot(x, x+3, 'm:', label='Dotted line', linewidth=2)
plt.plot(x, x+4, 'co-', label='Circles', markersize=2)
plt.plot(x, x+5, 's-', label='Squares', markersize=2)
plt.plot(x, x+6, '^-', label='Triangles', markersize=2)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Styles and Markers')
plt.legend()
plt.grid(visible=True, alpha=0.3)
plt.show()

# Common format string: '[color][marker][line]'
# Colors: b (blue), r (red), g (green), c (cyan), m (magneta), y (yellow), k (black)
# Markers: o (circle), s (square), ^ (triangle), * (star), + (plus), x (x)
# Lines: - (solid), -- (dashed), -. (dash-dot), : (dotted)
```
