```python
# Multiple lines on the same plot
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

plt.figure(figsize=(10,6)) # Length / height
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='sin(x) * cos(x)')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Multiple Functions')
plt.legend() # Show legend
plt.grid(visible=True, alpha=0.7) # alpha controls transparency
plt.show()
```

# 2. Bar Charts
**Bar charts are ideal for comparing categorical data or showing counts**


```python
# Simple bar chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]
config = {'xlabel': 'Categories', 'ylabel': 'Values', 'title': 'Simple Bar Chart'}

plt.figure(figsize=(5,3), dpi=200, facecolor='gray', edgecolor='red', linewidth=1.0)
plt.bar(categories, values, color='steelblue', edgecolor='red', linewidth=1.2)
plt.gca().set(**config) # It works nice instead of plt.xlabel('...') ; plt.ylabel('...') ; [:]
plt.gca().set_facecolor('lightgray') # whitesmoke / lightgray set axes background.
# Without explicitly called fig, ax = plt.subplots() ,plt.gcf() - sets figure background.
# But you have to have use `flat dictionaries`
plt.grid(visible=True, alpha=0.3, axis='y', color='green') # only horizontal grid lines
plt.show()
```

# 3. Horizontal Bar Charts

# 4. Grouped Bar Chart

```python
# Grouped bar chart
categories = ['Q1', 'Q2', 'Q3', 'Q4']
product_a = [20, 35, 30, 35]
product_b = [25, 32, 34, 20]
product_c = [30, 25, 30, 25]

x = np.arange(len(categories)) # Vector array([0, 1, 2 ,4])
width = 0.25 # Width of bars

plt.figure(figsize=(15,6))
# plt.bar(x ,height, width, ...) parameters:
# x - width: x-position (shift left by one bar width to position first group)
# product_a: height values for Product A bars
# width: width of each bar (0.25)
# label: text for legend
# color: bar fill color (hex code)
plt.bar(x=x-width, height=product_a, width=width, label='Product A', color='#FF6B6B')
# x: center position for second group (no shift)
plt.bar(x=x, height=product_b, width=width, label='Product B', color='#4ECDC4')
# x + width: x+position (shift right by one bar width to position third group)
plt.bar(x=x+width, height=product_c, width=width, label='Product C', color='#95E1D3')

plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.title('Qarterly Sales Comparision')
plt.xticks(ticks=x, labels=categories) # list of positions against list of labels
plt.legend() # shows legend :-)
plt.grid(visible=True, alpha=0.3, axis='y')
plt.show()
```

# 7. Subplots: Multiple Plots in One Figure
Subplots allow you t ocreate multiple plots in a single figure, which is great for compairng different visualizations.

```python
# Create 2x2 subplot grid
x = np.linspace(0, 10, 100)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Multiple Subplots', fontsize=16, fontweight='bold')

# Top left: line plot
axes[0, 0].plot(x, np.sin(x), 'b-', linewidth=2)
axes[0, 0].set_title('Sine Wave')
axes[0, 0].grid(visible=True, alpha=0.3)

# Top right: Cosine
axes[0, 1].plot(x, np.cos(x), 'r-', linewidth=2)
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].grid(visible=True, alpha=0.3)

# Bottom left: Scatter
np.random.seed(42)
x_scatter = np.random.randn(100)
y_scatter = np.random.randn(100)
axes[1, 0].scatter(x_scatter, y_scatter, alpha=0.6, c=x_scatter, cmap='viridis')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].grid(visible=True, alpha=0.3)

# Bottom right: Bar chart
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]
axes[1, 1].bar(x=categories, height=values, color='coral')
axes[1, 1].set_title('Bar Chart')

axes[1, 1].grid(visible=True, alpha=0.3, axis='y')

plt.tight_layout() # Adjust spacing between subplots
plt.show()
```

# Pandas

```python
# Create a sample DataFrame
np.random.seed(42)
dates = pd.date_range(start='2026-01-01', periods=100, freq='D') # Generate pd.DateTimeIndex
df = pd.DataFrame({
  'Date': dates,
  'Sales': np.random.randint(low=1000, high=5000, size=100) + np.sin(np.arange(100)) * 500,
  'Profit': np.random.randint(low=200, high=800, size=100) + np.cos(np.arange(100)) * 200,
  'Region': np.random.choice(['North', 'South', 'East', 'East', 'West'],100)
})
print("Sample DataFrame:")
print(df.head(n=5))
print(f"\nDataFrame shape: {df.shape}")
```

### Plot directly from DataFrame

```python
# Plot directly from DataFrame
plt.figure(figsize=(12,6), facecolor='gray')
plt.gca().set_facecolor('lightgreen')
plt.plot(df['Date'], df['Sales'], label='Sales', linewidth=2)
plt.plot(df['Date'], df['Profit'], label='Profit', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Amount ($)')
plt.title('Sales and Profit Over Time')
plt.legend()
plt.grid(visible=True, alpha=0.3)
plt.xticks(rotation=45) # Rotate x-axis label
plt.tight_layout() # Keeps layout tight :-)
plt.show()
```

## Pandas has his own plot() function

```python
# Using pandas plot method (even simpler!)
fig, axes = plt.subplots(2, 1, figsize=(12,10))

# Plot using pandas .plot() method
df.plot(x='Date', y='Sales', ax=axes[0], color='steelblue', linewidth=2)
axes[0].set_title('Sales Over Time (Pandas Plot)', fontsize=12)
axes[0].set_ylabel('Sales ($)')
axes[0].grid(visible=True, alpha=0.3)
axes[0].tick_params(axis='x', rotation=45)

# Bar chart by region
region_sales = df.groupby('Region')['Sales'].mean()
region_sales.plot(kind='bar', ax=axes[1], color='coral', edgecolor='black')
axes[1].set_title('Average Sales by Region', fontsize=12)
axes[1].set_ylabel('Average Sales ($)')
axes[1].set_xlabel('Region')
axes[1].grid(visible=True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

# 9. Advanced Customization: Annotations, Text, and Legends
Adding annotations and customizing text elements can make your plots more informative and profesional.

```python
# Adding annotations and text
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y, 'b-', linewidth=2, label='sin(x)')
# 'b-' means ax.plt(color='blue', linestyle='solid', linewidth=2) etc.

# Add text annotation
max_idx = np.argmax(y)
ax.annotate('Maximum', xy=(x[max_idx], y[max_idx]),
            xytext=(x[max_idx] + 1, y[max_idx] + 0.3),
            arrowprops=dict[str, any](arrowstyle='->', color='red', lw=2),
            fontsize=12, color='red', fontweight='bold'
            )

# Add text annotation_2
min_idx = np.argmin(y)
ax.annotate('Minimum', xy=(x[min_idx], y[min_idx]),
            xytext=(x[min_idx] - 3, y[min_idx] + 0.3),
            arrowprops=dict[str, any](arrowstyle='->', color='red', lw=2),
            fontsize=12, color='red', fontweight='bold'
            )

# Add text box
ax.text(0.5, 0.95, 'This is a sine wave',
        transform=ax.transAxes, fontsize=12,
        verticalalignment='top', bbox=dict[str, str | float]
        (boxstyle='round', facecolor='wheat', alpha=0.5))

# Add vertical and horizontal lines
ax.axvline(x=5, color='green', linestyle='--', alpha=0.7, label='x=5')
ax.axhline(y=0, color='black', linestyle='-', alpha=0.3)

ax.set_xlabel('X values', fontsize=12)
ax.set_ylabel('sin(x)', fontsize=12)
ax.set_title('Annotated Plot', fontsize=14, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(visible=True, alpha=0.3)
plt.show()
```

# Customizing Legends

```python
# Customizing Legends
x = np.linspace(0, 10, 100)

fig, ax = plt.subplots(figsize=(10,6),dpi=100)
ax.plot(x, np.sin(x), label='sin(x)', linewidth=2)
ax.plot(x, np.cos(x), label='cos(x)', linewidth=2)
ax.plot(x, np.sin(x) * np.cos(x), label='sin(x) * cos(x)', linewidth=2)

# Legend magic
ax.legend(loc='upper right', frameon=True, fancybox=True,
          shadow=True, fontsize=10, title='Functions')

ax.set_xlabel('X values')
ax.set_ylabel('Y values')
ax.set_title('Customized Legend')
ax.grid(visible=True, alpha=0.3)
plt.show()

# Common legend locations: 'upper right', 'upper left', 'lower right', 'lower left',
# 'center', 'best' (auto-detect best location)
```

# Pie Charts

```python
# Simple pie chart
categories = ['North', 'South', 'East', 'West']
sizes = [30, 25, 20, 25]
colors = ['#FF6B6B', '#4ECDC4', '#95E1D3', '#F38181']
explode = (0.1, 0, 0, 0) # Explode the first slice

fig, ax = plt.subplots(figsize=(8,8))
ax.pie(x=sizes, explode=explode, labels=categories, colors=colors,
       autopct='%1.1f%%', shadow=True, startangle=90)
ax.set_title('Sales Distribution by Region', fontsize=14, fontweight='bold')
plt.show()
```

# 12. Saving charts into files

```python
# Create a sample plot
x = np.linspace(start=0, stop=10, num=100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(10,6))
ax.plot(x, y, 'b-', linewidth=2)
ax.set_xlabel('X values')
ax.set_ylabel('Y values - sin(x)')
ax.set_title('Sample Plot for Saving')
ax.grid(visible=True, alpha=0.3)

# Save the figure
# Common formats: PNG (raster), PDF (vector), SVG (vector), EPS (vector)
fig.savefig(fname='my_plot.png', dpi=300, bbox_inches='tight')
fig.savefig('my_plot.pdf', bbox_inches='tight') # Vector format (scalable)

print("Figure saved as 'my_plot.png' and 'my_plot.pdf'")
print("\nCommon save parameters:")
print("- dpi: Resolution (dots per inch). Higher = better quality but larger file")
print("- bbox:_inches='tight': removes extra whitespace")
print("- format: Explicit format specification")
print("- facecolor: Background color")

plt.show()
```
